```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pickle
```

```python

df = pd.read_csv('updated_dataset_v3.csv')
df.head(2)
```

```python
df = df.drop(df.columns[0], axis=1)
df.head(5)
```

```python
#Cleaning any rows with empty values
df = df.dropna(axis=0, how='any')

df.head(5)
```

```python
# Converting categorical variables into numerical values
df.replace({'Yes': 1, 'No': 0}, inplace=True)
df.replace({'Never': 0, 'Rarely': 1, 'Occasionally': 2, 'Frequently': 3, 'Always': 4}, inplace=True)

# Filling in missing values
# df.fillna(df.mean(), inplace=True)

# Splitting the dataset into training and testing sets
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Training the model
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf.fit(X_train, y_train)

# Testing the model
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
# Making predictions
new_data = [[0, 0, 0, 0, 0, 0, 0, 2, 0, 0]]
prediction = clf.predict(new_data)
print('Prediction:', prediction)
```

```python
# Saving the model as pickle file
with open('addiction_model.pkl', 'wb') as f:
    pickle.dump(clf, f)
```
